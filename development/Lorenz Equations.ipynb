$$f(x) = \int_0^\infty f(\xi,x)d\xi$$

```python
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D

def lorenz_sys(q,t):
    x,y,z = q
    # sigma, rho and beta are global.
    f = [sigma * (y - x),
         rho*x - y - x*z,
         x*y - beta*z]
    return f

ic = [1.0, 2.0, 1.0]
t0 = 0.0
t1 = 100.0

sigma = 10.0
rho = 28.0
beta = 10.0/3

t = linspace(t0,t1,10000)
sol = odeint(lorenz_sys,ic,t)

fig = figure()
ax = Axes3D(fig)
ax.plot(sol[:,0], sol[:,1], sol[:,2])
xlabel('x')
ylabel('y')
show()
```
