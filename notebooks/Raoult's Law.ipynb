# Raoult's Law

Raoult's law says the partial pressure of each component of an ideal mixture of liquids is equal to the product of the mole fraction and the saturation pressure of the pure component. That is,

$$y_n P = x_n P^{sat}_n(T) $$

For an ideal gas, the total vapor pressure of the liquid is equal to the sum of partial pressures,

$$P = \sum_{n=1}^N x_n P^{sat}_n(T) $$

### Example: Acetone/Ethanol

In this example we estimate the vapor pressure of an acetone/ethanol mixture using Raoult's law, assuming the vapor pressure of the component species are given by Antoine's equation. The estimate is compared to experimentally measured data.

For a binary mixture of acetone ($A$) and ethanol ($B$)

$$P(T,x_A) = x_{A} P^{sat}_{A}(T) + (1-x_{A})P^{sat}_{B}(T)$$

```python
A = 'acetone'
B = 'ethanol'

# Antoine's equations
Psat[A] = lambda T: 10**(7.02447 - 1161.0/(T + 224))
Psat[B] = lambda T: 10**(8.04494 - 1554.3/(T + 222.65))

# Plot vapor pressure using Raoult's law
x = linspace(0,1)
P = lambda x: x*Psat[A](T) + (1-x)*Psat[B](T)
plot(x,map(lambda x: x*Psat[A](T) + (1-x)*Psat[B](T),x),'b')
plot(x,map(lambda x: x*Psat[A](T),x),'b--')
plot(x,map(lambda x: (1-x)*Psat[B](T),x),'b--')

# Experimental data of (P,x) observations
Px = array([\
    [11.679, 0.00000],\
    [14.999, 0.04220],\
    [16.585, 0.06730],\
    [19.358, 0.11300],\
    [22.571, 0.17870],\
    [24.811, 0.23610],\
    [25.585, 0.26650],\
    [27.384, 0.31280],\
    [28.371, 0.34700],\
    [31.037, 0.44580],\
    [33.037, 0.53720],\
    [35.370, 0.65480],\
    [37.584, 0.77210],\
    [38.890, 0.84740],\
    [40.130, 0.92520],\
    [41.317, 1.00000]])

# Convert kPa to mmHg
P = Px.T[0]*760/101.3

# Convert K to C
T = 305.15 - 273.15

# Extract measured composition
x = Px.T[1]

# Overlay plot of experimental data and label$$ 
plot(x,P,'ro')
xlim(0,1)
ylim(0,350)
xlabel('Mole fraction ' + A, fontsize = 14)
ylabel('Vapor Pressure [mmHg]', fontsize = 14)
title('Raoult\'s Law: ' + A + ' / ' + B + ' at {:.1f} deg C'.format(T), fontsize = 14)
legend(['Raoult\'s Law','','','Experimental'],'best')
grid()
```

For this example we observe a positive deviation of the experimentally measured vapor pressure from Raoult's law.

### Example: Vapor pressure of a binary mixture as function of composition

Using Raoult's law, plot the total vapor pressure of binary mixture of benzene and toluene as function of composition.

#### Solution

Using $x$ to denote the mole fraction of benzene, 

$$ P = x\, P^{sat}_{Benzene}(T) + (1-x)\, P^{sat}_{Toluene}(T)$$


```python
Tval = linspace(70,120,6)
x = linspace(0,1)

Psat = dict()
Psat['benzene'] = lambda T: 10**(6.90565 - 1211.033/(T + 220.790))
Psat['toluene'] = lambda T: 10**(6.95464 - 1344.8/(T + 219.48))

figure(figsize=(10,6))
for T in Tval:
    P = lambda x : x*Psat['benzene'](T) + (1-x)*Psat['toluene'](T)
    plot(x,map(P,x),linewidth=1.6)

xlabel('Mole Fraction Benzene',fontsize=14)
ylabel('Total Vapor Pressure',fontsize=14)
title('Vapor Pressure of a Benzene/Toluene Mixture at Several Temperatures',fontsize=14)
legend(Tval,'best')
grid()
```

```python
# Psat by Antoine's equation for selected compounds
Psat = dict()
Psat_Tmin = dict()
Psat_Tmax = dict()

Psat['benzene'] = lambda T: 10**(6.90565 - 1211.033/(T + 220.790))
Psat_Tmin['benzene'] = 8
Psat_Tmax['benzene'] = 103
    
Psat['ethanol'] = lambda T: 10**(8.04494 - 1554.3/(T + 222.65))
Psat_Tmin['ethanol'] = -2
Psat_Tmax['ethanol'] = 100
    
Psat['methanol'] = lambda T: \
    (T <= 65) and 10**(7.89750 - 1474.08/(T + 229.13)) or \
    (T > 65) and 10**(7.97328 - 1515.14/(T + 232.85))
Psat_Tmin['methanol'] = -14
Psat_Tmax['methanol'] = 110
    
Psat['toluene'] = lambda T: 10**(6.95464 - 1344.8/(T + 219.48))
Psat_Tmin['toluene'] = 6
Psat_Tmax['toluene'] = 137

Psat['water'] = lambda T: \
    (T <= 60) and 10**(8.10765 - 1750.286/(T + 235.0)) or \
    (T > 60) and 10**(7.96681 - 1668.21/(T + 228.0))
Psat_Tmin['water'] = 0
Psat_Tmax['water'] = 150


# Tsat functions
Tsat = dict()
Tsat_Pmin = dict()
Tsat_Pmax = dict()

from scipy.optimize import brentq as fzero

for s in Psat.keys():
    Tsat_Pmin[s] = Psat[s](Psat_Tmin[s])
    Tsat_Pmax[s] = Psat[s](Psat_Tmax[s])
    Tsat[s] = lambda P,s = s: fzero(lambda T: Psat[s](T)-P,Psat_Tmin[s],Psat_Tmax[s])
```
